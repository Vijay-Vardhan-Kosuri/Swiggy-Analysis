# Chart Insights

# Monthly sales trend

```python
print(df.columns)
print(df.head(2))
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Year&Month']=df['Order Date'].dt.to_period('M').astype(str)
print(df['Year&Month'])
monthly_sale=df.groupby('Year&Month')['Price (INR)'].sum().reset_index()
print(monthly_sale)
# Line trend
print(plt.figure())
print(plt.plot(monthly_sale['Year&Month'],monthly_sale['Price (INR)']))
print(plt.xticks(rotation=45))
print(plt.xlabel('Year&Month'))
print(plt.ylabel('Total Sales (INR)'))
print(plt.title('Monthly sales trend'))
plt.tight_layout()
plt.show()
```

# Daily Sales Trend

```python
df['Day_name']=pd.to_datetime(df['Order Date']).dt.day_name()
print(df['Day_name'])
df
```

```python
daily_sales=df.groupby('Day_name')['Price (INR)'].sum().reindex(['Monday','Tuesday',
                                                                  'Wednesday','Thursday','Friday','Saturday','Sunday'])

print(daily_sales)
# Bar plot
print(plt.figure(figsize=(10,5)))
print(plt.bar(daily_sales.index,daily_sales.values))
print(plt.xlabel('Day of the week'))
print(plt.ylabel('Total Sales (INR)'))
print(plt.title('Daily Sales'))
print(plt.xticks(rotation=25))
```

# Total sales by food type (veg vs non-veg)

```python
non_veg = [
    "chicken","duck","turkey","mutton","lamb","pork","beef",
    "fish","prawn","shrimp","crab","lobster","squid","octopus",
    "sausages","bacon","salami","egg"
    ]
df['Food_Category']=np.where(df['Dish Name'].str.lower().str.contains('|'
                                                                      .join(non_veg),na=False),'Non-Veg','Veg')
print(df['Food_Category'])
food_sales=(df.groupby('Food_Category')['Price (INR)'].sum()).reset_index()
print(food_sales)
print(food_sales.sum())
#pie chart
pio.renderers.default='browser'
fig=px.pie(
    food_sales,
    values="Price (INR)",
    names='Food_Category',
    hole=0.5,
    title="Revenue Distribution by Food Category",
)
fig.update_traces(
    textinfo="percent+label",
    pull=[0.05] * len(food_sales['Food_Category'].unique()),
    )
fig.update_layout(
    height=500,
    margin=dict(t=60,b=40,l=40,r=40),
)

```

# Total sales by state

```python
state_wise_sale = df.groupby('State')['Price (INR)'].sum().reset_index()
print(state_wise_sale)
# bar chart using plotly
pio.renderers.default='browser'
fig=px.bar(
    df.groupby('State',as_index=False)['Price (INR)'].sum().sort_values('Price (INR)',ascending=False),
    y='State',
    x='Price (INR)',
    orientation='h',
    title='Revenue by State (INR)'
    )
fig.update_layout(height=600,yaxis=dict(autorange='reversed'))
```

# Quarterly sales summary

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Quarterly']=df['Order Date'].dt.to_period('Q').astype(str)
print(df['Quarterly'])
quarterly_summary=(
    df.groupby('Quarterly',as_index=False)
.agg(
    Total_Sale=('Price (INR)', 'sum'),
    Avg_rating=('Rating', 'mean'),
    Total_orders=('Order Date','count')
)
.sort_values("Quarterly")
)
print(quarterly_summary)
```

# top 5 cities by sales

```python
pio.renderers.default='browser'

# --- Step 1: Data Cleaning ---
df['Price (INR)'] = pd.to_numeric(df['Price (INR)'], errors='coerce')
df = df.dropna(subset=['Price (INR)', 'City'])

# --- Step 2: Create Top 5 City Revenue Data ---
city_sale = (
    df.groupby("City", as_index=False)["Price (INR)"]
    .sum()
    .sort_values("Price (INR)", ascending=False)
    .head(5)
)

# --- Step 3: Sort for Better Visualization (small → big for horizontal bar) ---
city_sale = city_sale.sort_values("Price (INR)", ascending=True)

# --- Step 4: Create Plot ---
fig = px.bar(
    city_sale,
    x="Price (INR)",
    y="City",
    orientation='h',
    title='Top 5 Cities by Revenue (INR)',
    text="Price (INR)",   # show values on bars
    color="Price (INR)", # gradient effect
    color_continuous_scale="Reds"
)

# --- Step 5: Styling ---
fig.update_traces(
    texttemplate='%{text:.2s}',   # clean number format
    textposition='outside'
)

fig.update_layout(
    height=500,
    margin=dict(t=60, b=40, l=40, r=40),
    xaxis_title="Revenue (INR)",
    yaxis_title="City",
    title_x=0.5   # center title
)

# --- Step 6: Show Plot ---
```

# Weekly trend analysis

```python

pio.renderers.default = "browser"
# --- Step 1: Ensure datetime ---
df['Order Date'] = pd.to_datetime(df['Order Date'])

# --- Step 2: Create week column ---
df['week'] = df['Order Date'].dt.to_period('W').astype(str)

# --- Step 3: Aggregate ---
weekly_sales = (
    df.groupby('week', as_index=False)['Price (INR)']
    .sum()
)

# --- Step 4: Plot ---
fig = px.line(
    weekly_sales,
    x='week',
    y='Price (INR)',
    title='Weekly Sales Trend (INR)',
    markers=True
)

# --- Step 5: Styling ---
fig.update_layout(
    xaxis_title="Week",
    yaxis_title="Revenue (INR)",
    title_x=0.5
)

fig.show()
```
